```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/Carl/Desktop/Prodigy internship/PRODIGY_DS_01/content/population.csv',  encoding='utf-8')
df.head()
```

```python
df.info()
```

```python
df_clean = df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'], axis= 1)
```

```python
df_clean.info()
```

```python
df_clean.dropna(inplace=True)
```

```python
df_clean.head()
```

```python
selected_country = 'Ghana'

country_data = df_clean[df_clean['Country Name'] == selected_country]

years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()

population.shape
```

```python
plt.figure(figsize=(5, 3))
plt.bar(years, population, color='green')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population Variation of {selected_country} Over the Years')
```
